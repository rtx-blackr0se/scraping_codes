```python
import pandas as pd
import requests
import re
import numpy as np
from bs4 import BeautifulSoup
#9846500907
```

```python
response = requests.get('https://merolagani.com/LatestMarket.aspx')
soup = BeautifulSoup(response.text)
```

```python
#generating links

tr = soup.find_all('tr')

tr_list = tr[1:235]
    
links = []
for tr in tr_list:
    try:
        links.append(tr.a['href'])
    except:
        pass

linkz = []

for link in links:
    try:
        linkz.append("https://merolagani.com" + link)
    except:
        pass

len(linkz)
```

```python

responses = []

for link in linkz:
    print('getting link')
    responses.append(requests.get(link))
```

```python
soups = []

for response in responses:
    print('getting doc')
    soups.append(BeautifulSoup(response.text))
```

```python
len(soups)
```

```python
td = []

for soup in soups:
    td.append(soup.find_all('td'))
```

```python
rows = []

for t in td:
    rows.append(t[0:13]+t[-8:])

len(rows[0])
```
